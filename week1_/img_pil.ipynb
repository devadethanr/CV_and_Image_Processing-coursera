```python
from PIL import Image

def get_concat_h(img1, img2):
    # Ensure both images have the same height
    if img1.height != img2.height:
        raise ValueError("Both images must have the same height for horizontal concatenation.")

    # Create a new image with the combined width and height
    dst = Image.new('RGB', (img1.width + img2.width, img1.height))
    
    # Paste the original and quantized images side by side
    dst.paste(img1, (0, 0))
    dst.paste(img2, (img1.width, 0))

    return dst
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Assuming you have the updated get_concat_h function

# Load your grayscale image
image_gray = Image.open("resources/lenna.jpg").convert("L")

# Loop through different quantization levels
for n in range(3, 8):
    # Quantize the image
    quantized_image = image_gray.quantize(256 // 2**n)
    
    # Display the original and quantized images side by side
    plt.figure(figsize=(10, 10))
    plt.imshow(get_concat_h(image_gray, quantized_image))
    plt.title("256 Quantization Levels left vs {} Quantization Levels right".format(256 // 2**n))
    plt.show()
```
